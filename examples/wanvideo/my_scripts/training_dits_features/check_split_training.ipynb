```python
import torch as th
path = '/data/mint/SkelAg/training/'
a = th.load(f"{path}/heatmap_64x64_with_single_sample/0/0.pth", weights_only=False)
b = th.load(f"{path}/heatmap_64x64_with_single/0/3.pth", weights_only=False)
c = th.load(f"{path}/heatmap_64x64_with_single_sample_2/0/0.pth", weights_only=False)
d = th.load(f"{path}/heatmap_64x64_with_single_2/0/3.pth", weights_only=False)

lat_a = a[0][0]['latents']
lat_b = b[0][0]['latents']
lat_c = c[0][0]['latents']
lat_d = d[0][0]['latents']

print(a[0][0]['num_frames'])
print(c[0][0]['num_frames'])

print(a[0][1]['prompt'])
print(c[0][1]['prompt'])

print("Text encoding equality: ", th.allclose(a[0][1]['context'], c[0][1]['context']))

print(lat_a.shape, lat_b.shape, lat_c.shape, lat_d.shape)
print(th.allclose(lat_a, lat_b))
print(th.allclose(lat_a, lat_c))
print(th.allclose(lat_b, lat_d))
print(th.sum(lat_a - lat_b))
print(th.mean(lat_a - lat_b))
print(th.mean(lat_a - lat_c))
print(th.mean(lat_b - lat_d))
print(th.mean(lat_a), th.mean(lat_b))
```

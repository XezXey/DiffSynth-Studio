```python
import json
import numpy as np
import plotly

def vis_joints3d(j3d, fig=None):
    if fig is None:
        fig = plotly.graph_objects.Figure()

    x, y, z = j3d[:, 0], j3d[:, 1], j3d[:, 2]
    fig.add_trace(plotly.graph_objects.Scatter3d(
        x=x, y=y, z=z,
        mode='markers',
        marker=dict(
            size=4,
            color='blue',
        )
    ))
    fig.update_layout(scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    ))

    # Show in grid 5x5x5
    fig.update_layout(scene=dict(
        xaxis=dict(nticks=5, range=[-1, 1]),
        yaxis=dict(nticks=5, range=[-1, 1]),
        zaxis=dict(nticks=5, range=[-1, 2]),
    ))

    fig.show()

f = "/data/mint/Motion_Dataset/Mixamo/output_mixamo_720p/Walking/cam_0/skeleton_cam_0.json"
with open(f, 'r') as file:
    data = json.load(file)

print(data.keys())
print(data['kinematic_chains'])
rest_pose = [v for k, v in data['rest_offsets'].items()]
rest_pose = np.array(rest_pose)
j3d = np.array(data['joints_3d'])
print(j3d.shape)
print(rest_pose.shape)
# vis_joints3d(rest_pose)

j3d[:, :, 0:2] -= j3d[0:1, 0:1, 0:2]
vis_joints3d(j3d[0])
```
